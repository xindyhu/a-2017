```python
rows = [row for row in table_demographics[0].find_all('tr')]
```

```python
rows
```

```python
header_row = rows[0]
header_row
```

```python
columns = [col.get_text() for col in header_row.find_all('th') if col.get_text()]
```

```python
columns
```

```python
rem_nl = lambda s:s.replace("\n", "")
```

```python
columns = [rem_nl(col) for col in columns]
columns
```

```python
indexes = [rem_nl(row.find('th').get_text()) for row in rows[1:]]
```

```python
indexes
```

```python
def to_num(s):
    if s[-1] == '%':
        return int(s[:-1])
    elif s[-2:] == "%\n":
        return int(s[:-2])
    else:
        return None

values = []
for row in rows[1:]:
    for value in row.find_all("td"):
        values.append(to_num(value.get_text()))
values
```

```python
stacked_values_lists = [values[i::3] for i in range(len(columns))]
```

```python
stacked_values_lists
```

```python
def print_them(a, b, c):
    print("a", a, "b", b, "c", c)
print_them(1, 2, 3)
```
